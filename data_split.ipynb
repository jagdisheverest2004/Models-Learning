```python
import numpy as np
import model_package as mp

model_package = mp.ModelPackage
simple_model = mp.ModelPackage.SimpleLinearRegression
multi_model = mp.ModelPackage.MultipleLinearRegression
```

```python
multiple_data = np.array(
    [
        [2, 0, 50],
        [3, 1, 60],
        [4, 0, 70],
        [5, 2, 80],
        [6, 1, 90],
        [7, 0, 95],
        [8, 1, 98],
        [9, 2, 99],
        [10, 0, 100]
    ]
)

simple_data = np.array(
    [
        [4,52],
        [9,51],
        [8,50],
        [8,52],
        [7,70],
        [5,60],
        [6,65],
        [7,68],
        [6,63],
        [5,58]    
    ]
)


logistic_data = np.array(
    [
        [1, 0, 0],
        [2, 5, 0],
        [3, 5, 0],
        [4, 10, 1],
        [5, 10, 1],
        [6, 15, 1],
        [7, 15, 1],
        [8, 20, 1]
    ]
)
```

```python
train_simple_data, test_simple_data = model_package.split_dataset(data=simple_data, test_size=0.3, random_state=42)

slope, intercept = simple_model.simple_linear_regression(self=simple_model, train_data=train_simple_data)

print(f"Slope: {slope}, Intercept: {intercept}")
```

```python
sm_y_pred, sm_y_test = simple_model.simple_predict(self=simple_model, test_data=test_simple_data)

train_simple_data, test_simple_data = model_package.split_dataset(data=simple_data, test_size=0.3, random_state=42)

pk_y_pred, pk_y_test = model_package.predict(train_data=train_simple_data, test_data=test_simple_data)

print(f"Simple Model Predictions: {sm_y_pred}, Actual: {sm_y_test}")
print(f"Sklearn Predictions: {pk_y_pred}, Actual: {pk_y_test}")
```

```python
train_data, test_data = model_package.split_dataset(data=multiple_data, test_size=0.3, random_state=42)

slopes, intercept = multi_model.multiple_linear_regression(self=multi_model, train_data=train_data)

print(f"Slopes: {slopes}, Intercept: {intercept}")
```

```python
mp_y_pred, mp_y_test = multi_model.multiple_predict(self=multi_model, test_data=test_data)

pkm_y_pred, pkm_y_test = model_package.predict(train_data=train_data, test_data=test_data)

print(f"Multiple Model Predictions: {mp_y_pred}, Actual: {mp_y_test}")
print(f"Sklearn Predictions: {pkm_y_pred}, Actual: {pkm_y_test}")
```
